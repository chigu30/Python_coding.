```python
# Q2:What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

#Scatter plot is nothing but the place of intersection of data.

import matplotlib.pyplot as plt

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)

plt.xlabel('x axis')

plt.ylabel('y axis')

plt.title('random data scatter plot')
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
    
# The subplot() is used to plot multiple plots in one figure.

import numpy as np

import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5]) 
    
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])

y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])

y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])

y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)

plt.plot(x1,y1)

plt.title('line plot y1')

plt.subplot(2,2,2)

plt.plot(x2,y2)

plt.title('line plot y2')

plt.subplot(2,2,3)

plt.plot(x3,y3)

plt.title('line plot y3')

plt.subplot(2,2,4)

plt.plot(x4,y4)

plt.title('line plot y4')

plt.show
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

# Bar plot is a type of chart and they are used to compare data sets that are indipendent of each other.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.subplot(2,2,1)

plt.bar(company,profit)

plt.subplot(2,2,2)

plt.barh(company,profit)
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

# A box plot is a type of chart that shows the distribution of numerical data. 

# It displays the minimum,maximum,median and quartiles of a data set.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

data = [box1,box2]

plt.boxplot(data)

plt.xticks([1,2],['box1','box2'])

plt.ylabel('values')

plt.title('box plot')

plt.show
```

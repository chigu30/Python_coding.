```python
# Q1. What is multithreading in python? why is it used? Name the module used to handle threads in python.

# In python multithreading is the ability of program to execute multiple threads concurrently within a single process.

# Multithreading is used in Python to achieve parallelism, which can improve the performance of certain types of programs.

# Especially when dealing with tasks that involve I/O operations or can be executed independently.

# By utilizing multiple threads, the program can effectively utilize available system resources & perform tasks simultaneously.

# The module used to handle threads in Python is called 'threading'.

# It provides a high-level interface for creating and managing threads within a program.
```

```python
# Q2. Why threading module used? write the use of the following functions
# 1. activeCount()
# 2. currentThread()
# 3. enumerate()

# Threading module is used to implement multithreading in python.

# Which allows multiple threads execucation in a single process.

# The activeCount() function is used to return the number of Thread objects currently alive.

# It helps in determining the number of active threads in the current program.

# This can be useful in monitoring the progress or managing the synchronization of threads.

# The currentThread() function is used to retrieve the current Thread object corresponding to the caller's thread of execution.

# It returns an instance of the Thread class representing the currently executing thread.

# This function is useful when you need to access or manipulate properties of the current thread, 

# such as its name, identification number, or other attributes.

# The enumerate() function returns a list of all Thread objects currently alive.

# It can be used to obtain a list of all active threads within a program.

# Each thread in the list can then be inspected or manipulated as needed.

# This function is particularly useful when you want to iterate over all the threads 

# to perform some operation or gather information about each thread.
```

```python
# Q3. Explain the following functions
# 1. run()
# 2. start()
# 3. join()
# 4. isAlive()

# The run() function is a method defined in the Thread class in Python's threading module.

# It represents the code that will be executed by a thread when it is started.

# The start() function is also a method defined in the Thread class. It is used to start the execution of a thread.

# The join() function is another method of the Thread class.

# It allows a calling thread to wait for the completion of another thread before it continues its execution.

# The isAlive() function is a method of the Thread class that can be called on a thread object.

# It returns a boolean value indicating whether the thread is currently alive or not. 
```

```python
# Q4. Write a python program to create two threads. Thread one must print the list of squares and

# thread two must print the list of cubes.

import threading

def print_squares():
    squares = [i**2 for i in range(1, 11)]
    for square in squares:
        print(square)

def print_cubes():
    cubes = [i**3 for i in range(1, 11)]
    for cube in cubes:
        print(cube)

if __name__ == "__main__":
    # Create the first thread for printing squares
    thread1 = threading.Thread(target=print_squares)

    # Create the second thread for printing cubes
    thread2 = threading.Thread(target=print_cubes)

    # Start both threads
    thread1.start()
    thread2.start()

    # Wait for both threads to finish
    thread1.join()
    thread2.join()
```

```python
# Q6. Explain deadlocks and race conditions.

# DEADLOCKS :

# A deadlock is a situation where two or more processes or threads are unable to proceed because

# each is waiting for the other to release a resource.

# In other words, it's a state where processes are stuck in a circular waiting pattern, leading to a system-wide halt.

# Deadlocks typically occur due to a lack of proper synchronization or resource allocation.

# Race Conditions : 

# A race condition occurs when the behavior or outcome of a program depends on 

# the relative timing or interleaving of multiple threads or processes. 

# It arises when two or more threads/processes access shared data or resources concurrently, 

# and the final result depends on the order in which the operations are executed. 

# The race condition can lead to unpredictable and erroneous results.
```

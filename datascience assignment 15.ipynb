# Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight backend framework with minimal dependencies. Advantages of Flask Framework include:

    Easy to learn because of its simple and intuitive API.
    Flexible because it allows you to customize and extend the framework to suit your needs easily.
    Has a built-in development server.
    Has a fast debugger.
    Supports client-side session management.
    Well-documented framework with several sample examples.
    Minimalistic framework that gives a lot of freedom to its user in web development.
    Allows developers to have complete control over the codebase, which is small.
    Enables developers to choose the components they want for a specific application.
    Empowers developers to make decisions while selecting the components for development.

```python
# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()
```

# Q3. What is App routing in Flask? Why do we use app routes?

App routing is used to map the specific URL with the associated function that is intended to perform some task. The routing technique helps a user remember application URLs and is useful to access the desired page directly without having to navigate from the home page.

```python
# Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details: Company Name: ABC Corporation Location: India Contact Detail: 999-999-9999

from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def details():
    return 'Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999'

if __name__ == '__main__':
    app.run()
```

```python
# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Welcome to the homepage!'

@app.route('/user/<username>')
def user_profile(username):
    return f'Hello, {username}!'

@app.route('/product/<int:product_id>')
def product_details(product_id):
    return f'Details of product {product_id}'

if __name__ == '__main__':
    with app.test_request_context():
        
        homepage_url = url_for('index')
        user_profile_url = url_for('user_profile', username='John')
        product_details_url = url_for('product_details', product_id=42)

        print('Homepage URL:', homepage_url)
        print('User Profile URL:', user_profile_url)
        print('Product Details URL:', product_details_url)
```

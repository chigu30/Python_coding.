# Q1. Explain GET and POST methods.

The GET and POST methods are HTTP request methods used to send data to a server and retrieve a response. 

The GET method is used to retrieve data from a server. When a client sends a GET request, it includes parameters in the URL itself. These parameters are appended to the URL as query strings, separated by a question mark (?), and are typically in the format key=value.

The POST method is used to send data to a server to create or update a resource. Unlike GET, the POST request sends data in the request body rather than as part of the URL. The data can be in various formats such as JSON, XML, or form data.

# Q2. Why is request used in Flask?

In Flask, the request object is used to handle incoming HTTP requests from clients. It provides access to various properties and methods that allow developers to access and manipulate the data sent in the request.

# Q3. Why is redirect() used in Flask?

In Flask, the redirect() function is used to redirect the client's browser to a different URL. It is a helpful utility for handling HTTP redirects in web applications. 

# Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are used to separate the presentation logic from the application logic. Templates allow developers to define the structure and layout of web pages in a separate file, making it easier to manage and maintain the code.

The render_template() function in Flask is used to render a template file and generate a complete HTML page. It takes the name of the template file as a parameter and can also accept additional arguments to pass data to the template.

```python
# Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.


from flask import Flask, jsonify,request,render_template

app = Flask(__name__)

@app.route('/',methods = ['GET','POST'])
def home_page():
    return render_template("index.html")

@app.route('/postman_file',methods=["POST"])
def math_operation1():
    if(request.method == "POST"):
        opps = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if(opps == 'add'):
            r = num1 + num2
            result = 'The addition of  ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)
        if(opps == 'subtract'):
            r = num1 - num2
            result = 'The substraction of  ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)
        if(opps == 'multiply'):
            r = num1 * num2
            result = 'The multiplication of  ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)
        if(opps == 'divide'):
            r = num1 / num2
            result = 'The division of  ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)
        if(opps == 'square'):
            r = num1 ** num2
            result = 'The square of  ' + str(num1) + ' and ' + str(num2) + ' is ' + str(r)
        return jsonify(result)


if __name__=="__main__":
    app.run(host="0.0.0.0")


```

# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Five plots of seaborn library are :
    
    scatterplot - This is used to plot only the point of interaction.
    
    relplot - This is used to plot an graph using the plot data or in other language it shows relational data 
    
    catplot - This is catagorical plot. This is used to plot the catagory wise data.
    
    jointplot - This is used when we need both scatterplot and histogram plot in the same data.
    
    pairplot - This plots all the numerical data in one frame and compare each and every data.

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =

# "timepoint" and y = "signal" for different events and regions.

import seaborn as sns

fmri = sns.load_dataset('fmri')

sns.lineplot(x= fmri.timepoint,y= fmri.signal,data = fmri,style ='region',hue = 'event')
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 

# 'pclass', y = 'age' and y = 'fare'.

titanic = sns.load_dataset('titanic')

sns.boxplot(x = titanic.pclass,y = titanic.age)

sns.boxplot(x = titanic.pclass,y=titanic.fare)
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue

# parameter for the 'cut' column of the diamonds dataset.

diamonds = sns.load_dataset('diamonds')

sns.histplot(data = diamonds,x = 'price',hue = 'cut')
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column

# of the iris dataset.

iris = sns.load_dataset('iris')

sns.pairplot(hue = 'species',data = iris)
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

flights = sns.load_dataset('flights')

flight_pivot = flights.pivot(index='month',columns='year',values='passengers')

sns.heatmap(flight_pivot,annot=True,fmt='d')
```

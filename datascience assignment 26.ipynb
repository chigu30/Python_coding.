```python
# Q1. How can you create a Bokeh plot using Python code?

import bokeh.io

import bokeh.plotting

bokeh.io.output_notebook()

from bokeh.plotting import figure, output_file, show

graph = figure(title="Bokeh Line Graph")

x = [1, 2, 3, 4, 5]

y = [5, 4, 3, 2, 1]

graph.line(x, y)

output_file("line.html")

show(graph)
```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

# In Bokeh, glyphs are visual markers used to represent data points on a plot.

graph1 = figure(title="Bokeh Scatter Graph")

graph1.circle(x, y,size = 10)

output_file("line.html")

show(graph1)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

# bokeh provides variety of options to customize the appearance of bokeh plot including axes, title,legend.

# some examples are:

# title:

g2 = figure(title = 'this is how we can customize our title')

# axes:

g2.xaxis.axis_label = 'This is how we can customize the x axis'

g2.yaxis.axis_label = 'This is how we can customize the y axis'

# legend:

g2.circle(x,y,fill_color = 'red',legend_label = 'This is how we can customize the legend')

output_file('scattr.html')

show(g2)
```

```python
# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

# The Bokeh server is a Python library and server that allows you to create and deploy interactive plots that can be updated in 

# real-time.

from bokeh.plotting import figure, curdoc

from bokeh.models import Slider

from bokeh.layouts import column

p = figure()

r = p.line(x=[1, 2, 3], y=[1, 2, 3])

def update_plot(attr, old, new):
    r.data_source.data['y'] = [x**slider.value for x in [1, 2, 3]]

slider = Slider(start=1, end=5, value=1, step=1, title="Power")

slider.on_change('value', update_plot)

layout = column(p, slider)

app = curdoc()

app.add_root(layout)

```

```python
# Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

from bokeh.plotting import save

g2 = figure(title="Bokeh Scatter Graph")

g2.xaxis.axis_label = 'x axis'

g2.yaxis.axis_label = 'yaxis'

g2.circle(x, y,size = 10,fill_color = '#d2b48c')

save(g2,'chikun.html')

from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html')

if __name__ == '__main__':
    app.run()

@app.route('/')
def index():
    bokeh_plot_url = 'C:\\Users\\Lakxs\\chikun.html'
    return render_template('index.html', bokeh_plot_url=bokeh_plot_url)
```

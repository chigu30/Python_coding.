```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x = iris.sepal_length,y = iris.sepal_width)
```

```python
sns.scatterplot(x = iris.sepal_length,y = iris.petal_length)
```

```python
sns.scatterplot(x = iris.petal_length,y = iris.petal_width)
```

```python
sns.scatterplot(x = iris.petal_length,y = iris.sepal_length)
```

```python
sns.relplot(x = iris.sepal_length,y = iris.petal_length,data = iris,hue = 'species')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x = tips.total_bill,y = tips.tip)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip,data = tips,hue='smoker')
```

```python
sns.scatterplot(x = tips.total_bill,y = tips.sex)
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip,data = tips,hue = 'sex')
```

```python
sns.scatterplot(x = tips.total_bill,y = tips.day)
```

```python
sns.relplot(y = tips.total_bill,x = tips.day)
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip,data = tips,style='size')
```

```python
sns.relplot(x = tips.total_bill,y = tips.tip,data = tips,style='size',hue = 'time')
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = tips)
```

```python
sns.catplot(x = 'smoker',y = 'tip',data = tips)
```

```python
sns.jointplot(x = tips.total_bill,y = tips.tip)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(tips)
```
